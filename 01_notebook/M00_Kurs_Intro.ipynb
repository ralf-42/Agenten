# 2 | GenAI vs. Agenten - Der Unterschied
---


Generative KI und Agenten bauen aufeinander auf, sind aber nicht dasselbe.
| Ebene | Beschreibung | Typisches Ergebnis |
|---|---|---|
| **LLM-Call** | Ein Prompt, eine Antwort | Textantwort |
| **GenAI-App** | Prompt + UI + einfache Logik | Chatbot, Assistent |
| **KI-Agent** | LLM + Tool-Nutzung + Entscheidungslogik + Schleife | Mehrstufige Aufgabenerfuellung |


**Wann reicht GenAI ohne Agent?**     
- Reines Umformulieren, Zusammenfassen, Uebersetzen
- Keine externen Datenquellen oder Aktionen noetig   
  
**Wann lohnt ein Agent?**
- Mehrere Schritte mit Zwischenentscheidungen
- Externe Aktionen (APIs, Datenbanken, Dateien)
- Rueckfragen, Fehlerbehandlung und Zielverfolgung
