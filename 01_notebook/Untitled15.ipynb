```python
# ── 2. Client verbinden ────────────────────────────────────────────────────────
from gradio_client import Client

client = Client("Ralf42/simple_mcp")
```

```python
# Tool-Info abrufen
api_info = client.view_api(return_format="dict")
```

```python
# ── 3. Caesar ──────────────────────────────────────────────────────────────────
encrypted = client.predict(text="Hallo Welt", shift=3, api_name="/caesar_encrypt")
print(f"Verschlüsselt: {encrypted}")

decrypted = client.predict(text=encrypted, shift=3, api_name="/caesar_decrypt")
print(f"Entschlüsselt: {decrypted}")
```

```python
# ── 4. ROT13 ───────────────────────────────────────────────────────────────────
encoded = client.predict(text="Geheime Nachricht", api_name="/rot13")
print(f"ROT13:  {encoded}")
print(f"Zurück: {client.predict(text=encoded, api_name='/rot13')}")
```

```python
# ── 5. Vigenère ────────────────────────────────────────────────────────────────
enc = client.predict(text="Hallo Welt", key="GEHEIM", api_name="/vigenere_encrypt")
print(f"Verschlüsselt: {enc}")

dec = client.predict(text=enc, key="GEHEIM", api_name="/vigenere_decrypt")
print(f"Entschlüsselt: {dec}")
```

```python
# ── 6. Binär ───────────────────────────────────────────────────────────────────
binary = client.predict(text="Hi", api_name="/text_to_binary")
print(f"Binär: {binary}")

text = client.predict(binary=binary, api_name="/binary_to_text")
print(f"Text:  {text}")
```
