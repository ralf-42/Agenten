

## 📚 Notebook-Vorschlag zur Vertiefung (Erklärung & Code-Beispiele)

Dieser Vorschlag ergänzt die bereits geplanten 5 Haupt-Notebooks (`01_Agent_Grundlagen.ipynb` etc.) um spezialisierte, **konzeptfokussierte** Notebooks, die vor oder parallel zu den Hauptblöcken verwendet werden.

| Tag | Konzept-Fokus | Notebook-Name | Zweck & Inhalt |
| :--- | :--- | :--- | :--- |
| **Tag 1** | **Agent-Konzepte & TAO** | `10_Theorie_TAO_ReAct.ipynb` | **Erklärung:** Visualisiert den **Thought $\to$ Action $\to$ Observation (TAO)**-Zyklus. **Beispiel:** Ein simulierter TAO-Loop in Python (ohne LangChain), um die Logik des "Denkens" zu demonstrieren. |
| **Tag 2** | **LCEL & Pydantic** | `20_Beispiel_LCEL_Pydantic.ipynb` | **LCEL-Erklärung:** Detaillierte Code-Beispiele für `prompt \| llm \| parser` und die Verwendung von `RunnableParallel` zur Mappierung komplexer Inputs. **Structured Output:** Pydantic `BaseModel`-Definitionen und die Anwendung von `with_structured_output()` zur Entity-Extraction. |
| **Tag 3** | **Embeddings & Chunking** | `30_Theorie_RAG_Basis.ipynb` | **Konzept:** Visualisierung des **Embeddings**-Prozesses (Text $\to$ Vektor) und die Berechnung der Vektor-Ähnlichkeit. **Chunking-Experiment:** Code zur Demonstration der Auswirkungen von `chunk_size` und `overlap` auf ein Quelldokument. |
| **Tag 4** | **LangGraph State & Routing** | `40_Theorie_LangGraph_Basis.ipynb` | **State-Erklärung:** Simples Beispiel einer **State Machine** mit 3 Nodes (`Start`, `Zwischenschritt`, `Ende`) zur Veranschaulichung von `TypedDict` und `operator.add`. **Visualisierung:** Code zur Generierung eines Graph-Diagramms (Graphviz/Mermaid) **vor** der Kompilierung. |
| **Tag 5** | **Multi-Agent Patterns** | `50_Theorie_MultiAgent_Patterns.ipynb` | **Übersicht:** Code-Kommentare, die die **Supervisor**, **Hierarchical** und **Collaborative** Patterns gegenüberstellen. **Fokus:** Detailliertes, kommentiertes Beispiel für die **Supervisor-Routing-Logik** (ohne Worker-Logik), die `with_structured_output` nutzt. |

## 🎯 Didaktische Integration

Diese zusätzlichen Notebooks unterstützen die **EINMALIG-Regel** des Kurses, indem sie die theoretischen Erklärungen vom Haupt-Code trennen:

* **Vor der Praxis:** Die Theorie-Notebooks werden kurz vor den entsprechenden Blöcken als **Live-Demo** vom Trainer genutzt, um Konzepte **einmalig** zu erklären (z.B. `10_Theorie_TAO_ReAct.ipynb` vor Tag 1, Block 2).
* **Referenz:** Sie dienen den Teilnehmern als **nachschlagbares Code-Referenzdokument** für die Kernkonzepte, während sie in den Haupt-Notebooks die komplexen Agenten implementieren.
* **Visualisierung:** Sie sind die idealen Orte, um Diagramme und Grafiken (z.B. Vektorraum-Visualisierungen) direkt im Code zu erzeugen.
